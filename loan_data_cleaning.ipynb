```python
# replace the null values to the respect of its corresponding data types
```

```python
missing = df.isnull().sum()
missing = missing[missing > 0 ].sort_values(ascending = False)
print(missing)
```

```python
df.describe()
```

```python
#replacing null values in categorical columns
cat_cols = df.select_dtypes(include = "object").columns

for col in cat_cols :
    if df[col].isnull().sum() > 0 :
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# for numeric columns

num_cols = df.select_dtypes(include = ["int","float"]).columns
```

```python
for col in num_cols:
    if df[col].isnull().sum() > 0:
      df[col] =  df[col].fillna(df[col].median())
```

```python
df.info()
```

```python
# fix data types
df['LoanAmount'] = df['LoanAmount'].astype(float)

for col in cat_cols :
    df[col] = df[col].astype(str)
```

```python
#standardise text

for col in cat_cols:
   df[col] = df[col].str.strip().str.lower()
```

```python
#handle outliers
for col in['LoanAmount','ApplicantIncome','CoapplicantIncome']:
    upper = df[col].quantile(0.99)
                            
    df[col] = df[col].apply(lambda x :upper if x > upper else x )
```

```python
df.to_csv("cleaned_loan.csv",index = False)
```
